```julia
function hola(p)
    x = LinRange(0,5,1000)
    y = [pdf(Exponential(),i) for i in x]
    x1 = [-log(rand(1)[1]) for i in 1:p]
    y1 = [rand(1)[1] .* pdf(Exponential(),i) for i in x1]


    plot(x,y, label = "Función de probabilidad de la exp()")
    scatter!(x1,y1, label = "Algoritmo aceptación-rechazo")
end


```

```julia
hola(1000)

```
