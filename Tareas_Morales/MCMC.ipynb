```julia
using Distributions, LinearAlgebra, Plots
```

```julia
function MCMC(x0,theta,tao,n)
    error = rand(Normal(),n-1)
    X = vcat([x0],zeros(n-1))
    
    for i in 2:n
        X[i] = theta*X[i-1] + error[i-1]
    end

    return X
end
```

```julia
X = MCMC(10,0.1,1,10000) 
p1 = plot(X,label = "MCMC")
p2 = histogram(X,label = "MCMC", orientation=:horziontal, ylim = (minimum(X),maximum(X)))
cuadricula = grid(2,2, heights=(0.2,0.8), widths=(0.8,0.2))
p = Plots.Plot()
plot(p,p,p1,p2, legend=:none, layout=cuadricula)
```
